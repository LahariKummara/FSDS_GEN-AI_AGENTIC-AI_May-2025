# Import Required Libraries

```python
from gtts import gTTS
from IPython.display import Audio
```

# Define the Text You want to Convert

```python
text=''' 
Johny Johny Yes Papa
Eating Sugar No Papa
Telling lies No Papa
Open Your Mouth Ha ha ha!.Thank You.'''
```

# Convert Text to Speech Using gTTS

```python
text_to_speech=gTTS(text,lang='en',tld='com') # use 'en' for english language
text_to_speech.save('text_to_speech_gtts.wav') # Save the output as a .wav file
```

```python
from gtts.lang import tts_langs # list all languages that gTTS support
print(tts_langs())
```

# Play the Audio

This will display an audio player directly 

```python
sound_file='text_to_speech_gtts.wav'
Audio(sound_file,autoplay=False)
```

# Text to Speech Conversion using pyttsx3

Changing the gender using pyttsx3

```python
pip install pyttsx3
```

```python
import pyttsx3
from IPython.display import Audio

text='''
Johny Johny Yes Papa
Eating Sugar No Papa
Telling lies No Papa
Open Your Mouth Ha ha ha!.Thank You.'''

audio=pyttsx3.init()
audio.setProperty('rate',150)
audio.setProperty('volume',0.8)

# Change the Voices
voice=audio.getProperty('voices')

# 0 for male and 1 for female
audio.setProperty('voice',voice[0].id)  # for male voice
# audio.setProperty('voice',voice[1].id) # for female voice

# text-to-speech conversion
audio.say(text)

#save the audio file
audio.save_to_file(text,'test_male_voice.mp3')
#audio.save_to_file(text,'test_male_voice.mp3')

audio.runAndWait()


```
